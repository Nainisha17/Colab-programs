**Hardcoded Variables**

```python
# Version 1: Hardcoded Variables – FCFS Scheduling

processes = [1, 2, 3]
burst_time = [5, 3, 8]
ct = 0
print("Process  Burst Time  Completion Time")
for i in range(len(processes)):
    ct += burst_time[i]
    print(f"P{processes[i]}        {burst_time[i]}            {ct}")
```

```python
# Version 1: Hardcoded Variables – SJF Scheduling

burst_time = [6, 2, 8, 3]
burst_time.sort()
wt = 0
print("Process  Burst Time  Waiting Time")
for i, bt in enumerate(burst_time):
    print(f"P{i+1}        {bt}            {wt}")
    wt += bt
```

```python
# Version 1: Hardcoded Variables – FIFO Page Replacement

pages = [7, 0, 1, 2, 0, 3, 0, 4]
frames = []
capacity = 3
for page in pages:
    if page not in frames:
        if len(frames) < capacity:
            frames.append(page)
        else:
            frames.pop(0)
            frames.append(page)
    print("Frames:", frames)
```

**Keyboard Input**

```python
# Version 2: Keyboard Input-Addition

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
print("Sum:", x + y)
```

```python
# Version 2: Keyboard Input – SJF Scheduling

n = int(input("Enter number of processes: "))
burst_time = []

for i in range(n):
    burst_time.append(int(input(f"Enter burst time for P{i+1}: ")))

burst_time.sort()
waiting_time = 0

print("Process  Burst Time  Waiting Time")
for i in range(n):
    print(f"P{i+1}        {burst_time[i]}            {waiting_time}")
    waiting_time += burst_time[i]
```

```python
# Version 2: Keyboard Input – Factorial

n = int(input("Enter a number: "))
fact = 1

for i in range(1, n + 1):
    fact *= i

print("Factorial:", fact)
```

**Read from File – Single Input**

```python
# Version 3: Read from File (Single Input)-Even or Odd

with open("number.txt", "w") as f:
    f.write("9")

with open("number.txt", "r") as f:
    n = int(f.read())

if n % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Version 3: Read from File (Single Input)-CPU Burst Time Sum

with open("burst.txt", "w") as f:
    f.write("5 10 15")

with open("burst.txt", "r") as f:
    bursts = list(map(int, f.read().split()))

print("Total CPU Burst Time:", sum(bursts))
```

**Read from File – Multiple Inputs**

```python
# Version 4: Read from File (Multiple Inputs)-FCFS from File

with open("fcfs.txt", "w") as f:
    f.write("4\n3\n7")
with open("fcfs.txt", "r") as f:
    bt = [int(line) for line in f]
ct = 0
print("Process  Burst Time  Completion Time")
for i in range(len(bt)):
    ct += bt[i]
    print(f"P{i+1}        {bt[i]}            {ct}")
```

```python
# Version 4: Read from File (Multiple Inputs)-Deadlock Detection

with open("deadlock.txt", "w") as f:
    f.write("1 0\n0 1")

with open("deadlock.txt", "r") as f:
    alloc = [list(map(int, line.split())) for line in f]

deadlock = any(sum(row) > 1 for row in alloc)
print("Deadlock Detected" if deadlock else "No Deadlock")
```
